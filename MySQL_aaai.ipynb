```python
bws_score = bws[['text', 'translated', 'cnt', 'weakest_cnt', 'strongest_cnt']]
bws_score['score'] = (bws['strongest_cnt'] - bws['weakest_cnt']) / 8
bws_score
```

```python
from sklearn.preprocessing import minmax_scale
import pandas as pd
 
scaled_score = minmax_scale(bws_score[['score']], feature_range=(0,1))
 
bws_score['minmax_score'] = scaled_score
bws_score
```
