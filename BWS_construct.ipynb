```python
def check_common_tok(idx_list, thresh):
    tok_list = []
    for idx in idx_list:
        try:
            tok_list.append(set(word_tokenize(bws_dep.loc[idx].text)))
        except:
            print(idx)
            continue
    
    for idx in range(3):  # 0,1 
        for idx2 in range(1+idx, 3):   # 0: 1, 2, 1: 2
            if len(tok_list[idx] & tok_list[idx2]) >= thresh:   # 교집합 
                # print(f'{idx_list[idx]}, {idx_list[idx2]}간에 공통 토큰이 {thresh}개 이상입니다.')
                # print(tok_list[idx] & tok_list[idx2])
                return False 
    return True 
```

```python
def is_duplicated(prev_idx_list, idx_list):
    if idx_list in prev_idx_list:
        print(f'중복 질문지입니다.')
        return True
    else:
        return False
```

```python
short_idx = bws_short.index.tolist()
long_idx = bws_long.index.tolist()
for idx in range(1, len(bws_short) // 3 + 1):
    cnt = 0 
    idx_list = random.sample(short_idx, 3)
    while check_common_tok(idx_list, 5) == False:
        cnt += 1 
        # print(idx_list)
        idx_list = random.sample(short_idx, 3)
        if cnt == len(short_idx):
            break 
    bws_set[0]['Q' + str(idx)] = idx_list
    for idx2 in idx_list:
        short_idx.remove(idx2)

print(idx)
for idx in range(201, 200 + len(bws_long) // 3 + 1):
    cnt = 0
    idx_list = random.sample(long_idx, 3)
    while check_common_tok(idx_list, 8) == False:
        cnt += 1 
        idx_list = random.sample(long_idx, 3)
        if cnt == len(long_idx):
            break 
    bws_set[0]['Q' + str(idx)] = idx_list
    for idx2 in idx_list:
        long_idx.remove(idx2)

print(idx)
```

```python
for idx in range(1, 8):
    short_idx = bws_short.index.tolist()
    long_idx = bws_long.index.tolist()
    for idx2 in range(1, len(bws_short) // 3 + 1):
        cnt = 0 
        idx_list = random.sample(short_idx, 3)
        while check_common_tok(idx_list, 5) == False or is_duplicated(prev_idx_list, idx_list) == True:
            cnt += 1 
            idx_list = random.sample(short_idx, 3)
            if cnt == len(short_idx):
                break 
        bws_set[idx]['Q' + str(idx2)] = idx_list
        for idx3 in idx_list:
            short_idx.remove(idx3)

    print(idx2)
    for idx2 in range(201, 200 + len(bws_long) // 3 + 1):
        cnt = 0
        idx_list2 = random.sample(long_idx, 3)
        while check_common_tok(idx_list2, 8) == False or is_duplicated(prev_idx_list, idx_list2) == True:
            cnt += 1 
            idx_list2 = random.sample(long_idx, 3)
            if cnt == len(long_idx):
                break 
        bws_set[idx]['Q' + str(idx2)] = idx_list2
        for idx3 in idx_list2:
            long_idx.remove(idx3)
            
    prev_idx_list.extend(list(bws_set[idx-1].values()))
    print(idx2)
```
